```python
from autogluon.tabular import TabularDataset,TabularPredictor
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
data=pd.read_csv('./use.csv')
data=data.drop('Unnamed: 0',axis=1)
data.head()
```

```python
x_train,x_test=train_test_split(data,test_size=0.2)
data=TabularDataset(x_train)
data_1=TabularDataset(x_test)
```

```python
predictor=TabularPredictor(label='Sel',eval_metric='r2').fit(train_data=data)
```

```python
predictor.leaderboard(data_1,silent=True,extra_metrics=['rmse'])
```
