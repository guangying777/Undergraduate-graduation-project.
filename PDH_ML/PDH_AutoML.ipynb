```python
from autogluon.tabular import TabularDataset,TabularPredictor
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
data=pd.read_excel('./此PDH数据集.xlsx')
data.head()
```

```python
data=pd.concat([data.loc[:,'金属1':'WHSV(流量)'],data.loc[:,'Y']],axis=1)
data.head()
```

```python

x_train,x_test=train_test_split(data,test_size=0.3)
```

```python
data=TabularDataset(x_train)
```

```python
data_1=TabularDataset(x_test)
```

```python
predictor=TabularPredictor(label='Y',eval_metric='r2').fit(train_data=data,presets='best_quality')
```

```python
predictor.leaderboard(data_1,silent=True,extra_metrics=['r2'])
```

```python
predictor.feature_importance(data)
```

```python
data=pd.read_excel('./use.xlsx').drop(['Unnamed: 0'],axis=1)
data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test=train_test_split(data,test_size=0.3,random_state=1314)
```

```python
data=TabularDataset(x_train)
```

```python
data_1=TabularDataset(x_test)
```

```python
predictor=TabularPredictor(label='Y',eval_metric='r2',verbosity=1).fit(train_data=data)
```

```python
predictor.leaderboard(data_1,silent=True,extra_metrics=['rmse'])
```

```python
predictor.predict_multi()
```

```python
predictor.interpretable_models_summary()
```
