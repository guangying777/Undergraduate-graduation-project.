```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# 解决中文乱码问题
plt.rcParams["font.sans-serif"]=["SimHei"] 
plt.rcParams["axes.unicode_minus"]=False 
```

```python
data=pd.read_excel('.\B_dataset.xlsx')
data.head()
```

```python
data=data.loc[:,'金属1':'丙烯选择性']
data
```

```python
data.info()
```

```python
data.loc[:,'GHSV'].fillna(data.loc[:,'GHSV'].mean(),inplace=True)
data.info()
```

```python
data.to_csv('use_1.csv')
```

```python
data1 = list(set(pd.concat([data.iloc[:, 0], data.iloc[:, 3]])))
data1
```

```python
data_1 = pd.DataFrame(np.zeros((228, 9)), columns=data1)
data_1
```

```python
num_1 = 0
for i in data.iloc[:, 0]:
    if i in data1:
        index_2 = data1.index(i)
        index_1 = num_1
        data_1.iloc[index_1, index_2] = data.iloc[num_1, 1]
        num_1 = num_1 + 1
num_1 = 0
for i in data.iloc[:, 3]:
    if i in data1:
        index_2 = data1.index(i)
        index_1 = num_1
        data_1.iloc[index_1, index_2] = data.iloc[num_1, 4]
        num_1 = num_1 + 1
data_1 = data_1.dropna(axis=1)
```

```python
data_1
```

```python
data2 = list(set(data.iloc[:, 6]))
len(data2)
```

```python
data_2 = pd.DataFrame(np.zeros((228, 9)), columns=data2)
```

```python
num_1 = 0
for i in data.iloc[:, 6]:
    if i in data2:
        index_2 = data2.index(i)
        index_1 = num_1
        data_2.iloc[index_1, index_2] = data.iloc[num_1, 7]
        num_1 = num_1 + 1
data_2 = data_2.dropna(axis=1)
```

```python
data_2.head()
```

```python
data3 = list(set(data.iloc[:, 8]))
len(data3)
```

```python
data_3 = pd.DataFrame(np.zeros((228,11)), columns=data3)
```

```python
num_1 = 0
for i in data.iloc[:, 8]:
    if i in data3:
        index_2 = data3.index(i)
        index_1 = num_1
        data_3.iloc[index_1, index_2] = data.iloc[num_1, 9]
        num_1 = num_1 + 1
data_3 = data_3.dropna(axis=1)
```

```python
data_3
```

```python
data_use=pd.concat([data_1,data_2,data_3,data.loc[:,'C2H6':'丙烯选择性']],axis=1)
data_use
```

```python
data_use.fillna(0,inplace=True)
```

```python
data_use
```

```python
data_use.to_csv('use_4.csv')
```
