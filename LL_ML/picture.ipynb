```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pathlib import Path
model=['LGBM','CatBoost','Lasso','Ridge','SVR(rbf)','SVR(Linear)','RFR','ETR',
       'LGBM','CatBoost','Lasso','Ridge','SVR(rbf)','SVR(Linear)','RFR','ETR']
R2=[0.707091,0.721650,0.680310,0.695046,0.630373,0.680846,0.706832,0.699132,
      0.703954,0.704755,0.595153,0.649284,0.621153,0.613670,0.690612,0.671441]
MAE=[0.146650,0.135553,0.162511,0.152031,0.159406,0.153165,0.141870,0.134161,
       0.146954,0.144499,0.190410,0.171985,0.158350,0.179691,0.146887,0.141768]
Metal_label=['Yes','Yes','Yes','Yes','Yes','Yes','Yes','Yes',
             'No','No','No','No','No','No','No','No']
```

```python
p={'model' : model,
   'R2' : R2,
   'MAE' : MAE,
   'Metal label':Metal_label
   }
F=pd.DataFrame(p)
F
```

```python
sns.set(style='white')
sns.barplot(x='model', y='R2', hue='Metal label', data=F) 
plt.xlabel('model',fontsize=15,font=Path('..\TimesNewRoman.ttf'))
plt.ylabel('R$^2$',fontsize=15,font=Path('..\TimesNewRoman.ttf'))
plt.ylim((0,0.9))
plt.xticks(rotation=30,fontsize=12,fontname='Times New Roman')
```

```python
sns.set(style='white')
sns.barplot(x='model', y='MAE', hue='Metal label', data=F) 
plt.xlabel('model',fontsize=15,font=Path('..\TimesNewRoman.ttf'))
plt.ylabel('MAE',fontsize=15,font=Path('..\TimesNewRoman.ttf'))
plt.ylim((0,0.5))
plt.xticks(rotation=30,fontsize=12,fontname='Times New Roman')
```
